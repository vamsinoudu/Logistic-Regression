```python
car_purchase={'symboling', 'wheelbase', 'carlength',
        'boreratio', 'stroke', 'compressionratio',
       'horsepower', 'peakrpm', 'citympg', 'highwaympg'}
```

```python
clean_data=df.copy()
clean_data['price_update']=(clean_data['price']>16500)*1
clean_data['price_update'].head()
```

```python
y=clean_data[['price_update']].copy()
y.head()
```

```python
x=clean_data[car_purchase].copy()
x.columns
```

```python
y.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred))
```
